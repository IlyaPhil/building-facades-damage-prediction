```python
import pandas as pd
```

```python
damage_vectors = pd.read_csv('../data/interim/building_features.csv')

def is_number(x):
    if pd.isna(x):
        return False
    try:
        int(x)
        return True
    except (ValueError, TypeError):
        return False
    
yandex_buildings = damage_vectors[damage_vectors['building_name'].apply(is_number)]


yandex_buildings['building_name'] = yandex_buildings['building_name'].astype(int)

yandex_buildings.sort_values(by='building_name')
```
